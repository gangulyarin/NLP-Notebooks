```python
data = pd.read_csv("/kaggle/input/stumbleupon/train.tsv",delimiter="\t")
data.head()
```

```python
data["boilerplate"][12]
```

```python
import json

def extractFromBoilerPlate(boilerplate,param):
    boilerplate = json.loads(boilerplate)
    if param not in boilerplate.keys():
        return None
    else:
        return boilerplate[param]
```

```python
data["title"] = data["boilerplate"].apply(lambda x: extractFromBoilerPlate(x,"title"))
data["body"] = data["boilerplate"].apply(lambda x: extractFromBoilerPlate(x,"body"))
data["url"] = data["boilerplate"].apply(lambda x: extractFromBoilerPlate(x,"url"))
data.head()
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
data.info()
```

```python
data['title'].fillna(" ",inplace=True)
```

```python
max_len = 100
```

```python
# Tokenize the Texts and convert them into Sequences
tokenizer = Tokenizer()
tokenizer.fit_on_texts(data['title'].tolist())
text_seq = tokenizer.texts_to_sequences(data['title'])
text_seq = pad_sequences(text_seq,maxlen=max_len)
```

```python
# Donwload Glove
import urllib.request

urllib.request.urlretrieve('https://nlp.stanford.edu/data/glove.6B.zip','glove.6B.zip')
```

```python
embed = dict()
f = open('glove.6B.300d.txt')

for line in f:
    tokens = line.split(" ")
    word = tokens[0]
    vector = np.asarray(tokens[1:],dtype="float")
    embed[word]=vector

embed["travel"]
```

```python
# The vocabulary
tokenizer.word_index
```

```python
vocab = tokenizer.word_index
```

```python
# Create a weight matrix for the word embeddings
embedding = np.zeros((len(vocab),300))

for i,word in enumerate(vocab):
    if word in embed.keys():
        vector = embed[word]
        embedding[i] = vector
```

```python
embedding.shape
```

```python
from sklearn.model_selection import train_test_split

train_X, test_X, train_y, test_y = train_test_split(text_seq,data['label'])
```

```python
train_X.shape
```

```python
# Build Model

model = tf.keras.models.Sequential()

model.add(tf.keras.layers.Embedding(len(vocab),300,embeddings_initializer=keras.initializers.Constant(embedding),
                                     trainable=False))
model.add(tf.keras.layers.LSTM(128,return_sequences=True))
#Global Maxpooling
model.add(tf.keras.layers.GlobalMaxPooling1D())

model.add(tf.keras.layers.Dense(64,activation="relu"))
model.add(tf.keras.layers.Dense(1,activation="sigmoid"))

#Add loss function, metrics, optimizer
model.compile(optimizer='adam', loss='binary_crossentropy',metrics=["acc"]) 

model.summary()
```

```python
model.fit(np.array(train_X),train_y,epochs=5)
```
