```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
```

```python
! pip install datasets
```

```python
from datasets import Dataset

train_ds = Dataset.from_pandas(data)
```

```python
type(train_ds)
```

```python
train_ds
```

```python
def preprocess(dataset):
    return tokenizer(dataset["question_text"],return_tensors="np",padding=True)

tokenized_dataset = preprocess(train_ds)
```

```python
labels = np.array(train_ds["target"])
```

```python
tokenized_dataset = dict(tokenized_dataset)
```

```python
from transformers import TFAutoModelForSequenceClassification

model = TFAutoModelForSequenceClassification.from_pretrained("bert-base-uncased")

model.compile(optimizer="adam")
```

```python
model.fit(tokenized_dataset,labels)
```
